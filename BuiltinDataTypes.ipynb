```python
string = input()
i ,j =0, len(string)-1
while(j>=i):
    if string[i]== string[j]:
        i+=1
        j-=1
    else:
        break
if j<=i :
    print(f"given is palindrome : {string} ")
else :
    print(f"not a palindrome : {string}")
```
